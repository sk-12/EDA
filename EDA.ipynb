```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('iris_dataset.csv')
```

```
print(df.head())
df['species'].unique()
```

####Univariate analysis


```

data_setosa=df.loc[df['species']=='setosa']
data_versicolor=df.loc[df['species']=='versicolor']
data_virginica=df.loc[df['species']=='virginica']

```

```
plt.plot(data_setosa['petal_length'],np.zeros_like(data_setosa['species']),'o')
plt.plot(data_versicolor['petal_length'],np.zeros_like(data_versicolor['species']),'o')
plt.plot(data_virginica['petal_length'],np.zeros_like(data_virginica['species']),'o')
plt.show()
plt.savefig('uivariate analysis.png')
```

####Bivariate analysis


```

sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'petal_length','sepal_length').add_legend()
plt.show()
plt.savefig('bivariate analysis.png')
```

####Multivariate Analysis

```
sns.pairplot(df,hue='species',size=5)
plt.savefig('multivariate analysis.png')
```
